# ML Homework 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing Test and Training Data 

```python
# Reference:
# https://web.stanford.edu/~hastie/ElemStatLearn/
df = pd.read_csv("/Users/schand/Documents/ML Homework 1/zip.train.csv",header = None)
X_train = df.iloc[:,1:257].values
y_train = df.iloc[:,0].values
df.head()
#Test data
test = pd.read_csv("/Users/schand/Documents/ML Homework 1/zip.test.csv",header=None)
X_test = test.iloc[:,1:257].values
y_test = test.iloc[:,0].values
test.head()
```

# Data with Label 2

```python
# Making output binary for linear regression model
y_train2 = [1 if item == 2 else 0 for item in y_train]
y_test2 = [1 if item == 2 else 0 for item in y_test]
```

- Linear Regression

```python
#Linear Regression Model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score

linear_regression_model = LinearRegression()
linear_regression_model.fit(X_train, y_train2)


def predict(X):
    return 1*(linear_regression_model.predict(X)> 0.5)

#Training and Test error
train_error_rate = 1 - accuracy_score(y_train2, predict(X_train))
test_error_rate = 1 - accuracy_score(y_test2, predict(X_test))
print("Training error for label 2 data set using linear regression model is " ,train_error_rate)
print("Test error for label 2 data set using linear regression model is " ,test_error_rate)
```

- Nearest Neighbor

```python
#Nearest Neighbor
from sklearn.neighbors import KNeighborsClassifier

# K = 15
neighbors15_classifier = KNeighborsClassifier(n_neighbors=15)
neighbors15_classifier.fit(X_train, y_train2)

def predict(X):
    return 1*(neighbors15_classifier.predict(X)> 0.5)

train_error_rate = 1 - accuracy_score(y_train2, predict(X_train))
test_error_rate = 1 - accuracy_score(y_test2, predict(X_test))
print("Training error for label 2 data set using k = 15 nearest neighbour is " ,train_error_rate)
print("Test error for label 2 data set using using k = 15 nearest neighbour is " ,test_error_rate)

# K = 7
neighbors7_classifier = KNeighborsClassifier(n_neighbors=7)
neighbors7_classifier.fit(X_train, y_train2)

def predict(X):
    return 1*(neighbors7_classifier.predict(X)> 0.5)

train_error_rate = 1 - accuracy_score(y_train2, predict(X_train))
test_error_rate = 1 - accuracy_score(y_test2, predict(X_test))
print("Training error for label 2 data set using k = 7 nearest neighbour is " ,train_error_rate)
print("Test error for label 2 data set using using k = 7 nearest neighbour is " ,test_error_rate)

# K = 5

neighbors5_classifier = KNeighborsClassifier(n_neighbors=5)
neighbors5_classifier.fit(X_train, y_train2)

def predict(X):
    return 1*(neighbors5_classifier.predict(X)> 0.5)

train_error_rate = 1 - accuracy_score(y_train2, predict(X_train))
test_error_rate = 1 - accuracy_score(y_test2, predict(X_test))
print("Training error for label 2 data set using k = 5 nearest neighbour is " ,train_error_rate)
print("Test error for label 2 data set using using k = 5 nearest neighbour is " ,test_error_rate)

#K = 3

neighbors3_classifier = KNeighborsClassifier(n_neighbors=3)
neighbors3_classifier.fit(X_train, y_train2)

def predict(X):
    return 1*(neighbors3_classifier.predict(X)> 0.5)

train_error_rate = 1 - accuracy_score(y_train2, predict(X_train))
test_error_rate = 1 - accuracy_score(y_test2, predict(X_test))
print("Training error for label 2 data set using k = 3 nearest neighbour is " ,train_error_rate)
print("Test error for label 2 data set using using k = 3 nearest neighbour is " ,test_error_rate)

# K = 1

neighbors1_classifier = KNeighborsClassifier(n_neighbors=1)
neighbors1_classifier.fit(X_train, y_train2)

def predict(X):
    return 1*(neighbors1_classifier.predict(X)> 0.5)

train_error_rate = 1 - accuracy_score(y_train2, predict(X_train))
test_error_rate = 1 - accuracy_score(y_test2, predict(X_test))
print("Training error for label 2 data set using k = 1 nearest neighbour is " ,train_error_rate)
print("Test error for label 2 data set using using k = 1 nearest neighbour is " ,test_error_rate)
```
